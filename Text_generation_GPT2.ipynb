```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

# Load Model

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
model = GPT2LMHeadModel.from_pretrained("gpt2-large", pad_token_id=tokenizer.eos_token_id)
```

```python
tokenizer.decode(tokenizer.eos_token_id)
```

# Tokenize Sentences

```python
sentence = "No-Code Needs To Adapt To Specialists"
input_id = tokenizer.encode(sentence,return_tensors='pt')
```

```python
input_id
```

```python
tokenizer.decode(input_id[0][0])
```

```python
tokenizer.decode(input_id[0][2])
```

# Generate Text

```python
output = model.generate(input_id, max_length=500, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
output
```

```python
tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
text = tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
with open('blogposticecrem.text','w') as f:
  f.write(text)
```
